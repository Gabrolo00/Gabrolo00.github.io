```python
#NEURONE TRAINING con forward, loss, backward, update, zero_grad
import torch

# Dati: il neurone deve imparare che y = 3x
x = torch.tensor([1., 2., 3., 4., 5.])
y_target = torch.tensor([3., 6., 9., 12., 15.])

# Peso iniziale casuale e bias
w = torch.tensor(0.5, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

lr = 0.01  # learning rate

for epoch in range(1000):
    # 1. FORWARD: calcola la predizione
    y_pred = w * x + b  # riempi tu

    # 2. LOSS: calcola l'errore (usa il mean squared error)
    loss =  ((y_pred - y_target) ** 2).mean() # riempi tu

    # 3. BACKWARD: calcola i gradienti
    loss.backward()


    # 4. UPDATE: aggiorna w e b (senza tracciare i gradienti)
    with torch.no_grad():
        w -= lr * w.grad
        b -= lr * b.grad

    # 5. AZZERA i gradienti per il prossimo giro
    w.grad.zero_()
    b.grad.zero_()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: loss={loss.item():.4f}, w={w.item():.4f}, b={b.item():.4f}")
```

```python
#TRAINING CON OPTIMIZER

#NEURONE TRAINING con forward, loss, backward, update, zero_grad
import torch

# Dati: il neurone deve imparare che y = 3x
x = torch.tensor([1., 2., 3., 4., 5.])
y_target = torch.tensor([3., 6., 9., 12., 15.])

# Peso iniziale casuale e bias
w = torch.tensor(0.5, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

lr = 0.01  # learning rate

optimizer = torch.optim.SGD([w, b], lr=lr)

for epoch in range(100):
    # 1. FORWARD: calcola la predizione
    y_pred = w * x + b  # riempi tu

    # 2. LOSS: calcola l'errore (usa il mean squared error)
    loss =  ((y_pred - y_target) ** 2).mean() # riempi tu

    # 3. BACKWARD: calcola i gradienti
    loss.backward()


    optimizer.step()     # aggiorna i pesi
    optimizer.zero_grad() # azzera i gradienti

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: loss={loss.item():.4f}, w={w.item():.4f}, b={b.item():.4f}")
```

```python
#Rete single layer

import torch.nn as nn

class SingleNeuron(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(5,1)

  def forward(self, x):
    return self.linear(x)

x = torch.tensor([1., 2., 3., 4., 5.]).reshape(-1, 5)
y_target = torch.tensor([3., 6., 9., 12., 15.]).reshape(-1, 1)



model = SingleNeuron()
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

print(x.shape)
print(y_target.shape)

for epoch in range(100):
  y_pred = model(x)
  loss = criterion(y_pred, y_target)
  loss.backward()
  optimizer.step()
  optimizer.zero_grad()

  if epoch % 10 == 0:
      print(f"Epoch {epoch}: loss={loss.item():.4f}")

print(model.linear.weight)
print(model.linear.bias)
```

```python
#MNIST | 3 layer + RELU + dropout + train + test

import torch
import torch.nn as nn
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

train_data = datasets.MNIST(root='./data', train=True, download=True,
                            transform=transforms.ToTensor())
test_data = datasets.MNIST(root='./data', train=False, download=True,
                           transform=transforms.ToTensor())

# Prendi una singola immagine e guardala
image, label = train_data[0]
print(image.shape)
print(label)

# Le immagini in PyTorch seguono il formato [C, H, W]: canali, altezza, larghezza.

# 1 canale → scala di grigi (come nel tuo caso, probabilmente MNIST)
# 3 canali → RGB (rosso, verde, blu)
# 4 canali → RGBA (RGB + trasparenza)


class MLP(nn.Module):
  def __init__(self):
    super().__init__()
    self.layer1 = nn.Linear(784,256)
    self.layer2 = nn.Linear(256,128)
    self.layer3 = nn.Linear(128,10)
    self.relu = nn.ReLU()
    self.Dropout = nn.Dropout(0.2)

  def forward(self,x):
    x = x.view(-1,784)
    x = self.relu(self.layer1(x))
    x = self.Dropout(x)
    x = self.relu(self.layer2(x))
    x = self.Dropout(x)
    x = self.layer3(x)
    return x


train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)

model = MLP()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)

for epoch in range(5):
    model.train()
    for images, labels in train_loader:
        # scrivi tu i 4 passi: forward, loss, backward, update+zero_grad
        y_pred = model(images)
        loss = criterion(y_pred, labels)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()



    if epoch % 1 == 0:
        print(f"Epoch {epoch}: loss={loss.item():.4f}")


```

```python
model.eval()
correct = 0
total = 0

with torch.no_grad():
  for images, labels in train_loader:
      outputs = model(images)
      predicted = torch.argmax(outputs, dim=1)
      total += labels.size(0)
      correct += (predicted == labels).sum().item()



print(f"Accuracy: {correct/total*100:.2f}%")
```
